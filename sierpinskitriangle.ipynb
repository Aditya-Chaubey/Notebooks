```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000 # capital variable shows that N is constant
# Intializing x and y vectors
sx = np.zeros(N)
sy = np.zeros(N)
# x and y vectors that contain 10,000 zeroes
```

```python
for i in range(1, N):
  # generate a random number from {1, 2, 3}
  k = np.random.randint(1, 4)

  # update x and y points
  sx[i] = sx[i-1]/2 + k-1
  sy[i] = sy[i-1]/2

  if k==2:
    sy[i] += 2
```

```python
plt.plot(sx, sy, 'k.', markersize = 1)
plt.title('Sierpinski Triangle')
plt.axis('off')
plt.show()
```
