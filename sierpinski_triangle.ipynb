```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000

sx = np.zeros(N)
sy = np.zeros(N)

for i in range (1,N):
  k = np.random.randint(1,4)
  sx[i] = (0.5 * sx[i-1]) + k - 1
  sy[i] = 0.5 * sy[i-1]

  if k == 2:
    sy[i] += 2


plt.plot(sx, sy, 'k.', markersize = 1)
plt.title("Sierpinski Triangle")
plt.axis("off")
plt.show() 
```
